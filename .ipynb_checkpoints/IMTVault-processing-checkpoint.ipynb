```python
import pandas as pd

examples = pd.read_csv("./imtvault/cldf/examples.csv")
examples
```

```python
skipped = 0
used = 0

for name, group in examples.groupby(by="Language_ID"):
    # Write a separate file for each language
    with open("./imtvault_parsed/" + name + ".txt", "w") as outfile:
        for _, row in group.iterrows():
            transcr = row["Analyzed_Word"]
            glosses = row["Gloss"]
            
            if type(transcr) is float or type(glosses) is float:
                # Bad row, skip
                skipped += 1
                continue
            else:
                used += 1
            
            transcr = transcr.replace("\t", " ")
            # transcr = transcr.replace("-", "")
            # transcr = transcr.replace("=", "")
            transcr = transcr.replace("?", " ?")
            transcr = transcr.replace(".", " .")
            transcr = transcr.replace("!", " !")
            transcr = transcr.replace(",", " ,")
            
            outfile.write("\\t " + transcr)
            glosses = glosses.replace("\t", " ")
            outfile.write("\n\\g " + glosses)
            
            translation = row["Translated_Text"]
            # Some rows might not have translation, that's fine
            if not (type(translation) is float and pd.isna(translation)):
                outfile.write("\n\\l " + translation + "\n\n")
                
print(f"Processed {used} rows and skipped {skipped}")
```
