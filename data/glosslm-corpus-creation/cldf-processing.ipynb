```python
import pandas as pd

# Load APICS
apics = pd.read_csv("../data/apics/cldf/examples.csv")
apics_langs = pd.read_csv("../data/apics/cldf/languages.csv")

apics = pd.merge(apics, apics_langs, left_on="Language_ID", right_on="ID", how="left")
apics = apics[['Analyzed_Word', 'Gloss', 'Translated_Text', 'Glottocode', 'ID_x']]
apics = apics[pd.notnull(apics['Analyzed_Word'])]
apics = apics[pd.notnull(apics['Gloss'])]
apics = apics.rename(columns={"ID_x": "id", 
                                    "Analyzed_Word": "transcription", 
                                    "Gloss": "glosses", 
                                    "Translated_Text": "translation",
                                    "Glottocode": "glottocode"})
apics['source'] = 'apics'
apics['metalang_glottocode'] = ''
apics['id'] = 'apics_' + apics['id'].astype(str)
apics
```

```python
import pandas as pd

# Load UraTyp
uratyp = pd.read_csv("../data/uratyp/examples.csv")
uratyp_langs = pd.read_csv("../data/uratyp/languages.csv")

uratyp = pd.merge(uratyp, uratyp_langs, left_on="Language_ID", right_on="ID", how="left")
uratyp = uratyp[['Analyzed_Word', 'Gloss', 'Translated_Text', 'Glottocode', 'ID_x']]
uratyp = uratyp[pd.notnull(uratyp['Analyzed_Word'])]
uratyp = uratyp[pd.notnull(uratyp['Gloss'])]
uratyp = uratyp.rename(columns={"ID_x": "id", 
                                    "Analyzed_Word": "transcription", 
                                    "Gloss": "glosses", 
                                    "Translated_Text": "translation",
                                    "Glottocode": "glottocode"})
uratyp['source'] = 'uratyp'
uratyp['metalang_glottocode'] = ''
uratyp['id'] = 'uratyp_' + uratyp['id'].astype(str)
uratyp
```

```python
imtvault = pd.read_csv("../data/imtvault_1_1/cldf/examples.csv")
imtvault = imtvault.drop(columns=['Primary_Text', 'Corpus_Reference', 'Abbreviations', 'Source', 'Contribution_ID', 'Comment', "Language_Name"])
imtvault = imtvault.rename(columns={"Language_ID": "glottocode", 
                                    "Analyzed_Word": "transcription", 
                                    "Gloss": "glosses", 
                                    "Translated_Text": "translation",
                                    "Meta_Language_ID": "metalang_glottocode",
                                   "ID": "id"})
imtvault['source'] = 'imtvault'
imtvault
```

```python
examples = pd.concat([uratyp, apics, imtvault])

original_transcriptions = examples['transcription'].str.replace('\t', ' ') 

examples['transcription'] = original_transcriptions.str.replace(r"(\w)\?", r"\1 ?", regex=True) \
                                                   .str.replace(r"(\w)\.", r"\1 .", regex=True) \
                                                   .str.replace(r"(\w)\!", r"\1 !", regex=True) \
                                                   .str.replace(r"(\w)\,", r"\1 ,", regex=True) \
                                                   .str.replace("\-(\s|$)", " ", regex=True)


examples['glosses'] = examples['glosses'] \
            .str.replace("\t", " ") \
            .str.replace("\-(\s|$)", " ", regex=True) \
            .str.replace(r"(\w)\.(\s|$)", r"\1 . ", regex=True) \
            .str.replace(r"(\w)\!(\s|$)", r"\1 ! ", regex=True) \
            .str.replace(r"(\w)\?(\s|$)", r"\1 ? ", regex=True)



# examples["is_segmented"] = examples["LGR_Conformance_Level"] == "LGRConformance.MORPHEME_ALIGNED"
examples["is_segmented"] = examples["transcription"].str.contains("-")
examples.loc[examples["is_segmented"] == False, "is_segmented"] = ""
examples.loc[examples["is_segmented"] == True, "is_segmented"] = "yes"

examples
```

```python
import langid
from tqdm import tqdm

tqdm.pandas()

iso_to_glotto = {
    "en": "stan1293",
    "fr": "stan1290",
    "de": "stan1295",
    "pt": "port1283",
    "zh": "mand1415",
    "es": "stan1288"
}

bad_transl = []

def compare_pred_lang(row):
    if not isinstance(row['translation'], str):
        bad_transl.append(row)
        row['metalang_glottocode'] = ''
        row['translation'] = ''
        return row
        
    pred_lang = langid.classify(row['translation'])[0]
    if pred_lang not in iso_to_glotto:
        # Unknown language
        bad_transl.append(row)
        row['metalang_glottocode'] = ''
        row['translation'] = ''
    elif iso_to_glotto[pred_lang] != row['metalang_glottocode']:
        if row['metalang_glottocode'] is not None and len(row['metalang_glottocode']) > 0:
            # We have a mismatch! Clear everything and run
            bad_transl.append(row)
            row['metalang_glottocode'] = ''
            row['translation'] = ''
        else:
            # Didn't have metalang, so let's go with it
            row['metalang_glottocode'] = iso_to_glotto[pred_lang]
    return row
    

# Check if predicted lang matches metalang
examples = examples.progress_apply(compare_pred_lang, axis=1)

print(f"Found {len(bad_transl)} broken translation instances")
examples
```
